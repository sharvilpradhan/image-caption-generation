```
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.bleu_score import SmoothingFunction
import pickle
import string
```

```
def bleu_score(can,ref):
  score1 = sentence_bleu(ref, can, weights=(1,0,0,0))
  score2 = sentence_bleu(ref, can, weights=(0.5,0.5,0,0))
  score3 = sentence_bleu(ref, can, weights=(0.33,0.33,0.33,0))
  score4 = sentence_bleu(ref, can, weights=(0.25,0.25,0.25,0.25))

  return score1,score2,score3,score4
```

```
with open('/content/caption_dict.pickle','rb') as input:
  cap_dict = pickle.load(input)
```

```
print(cap_dict)
```

```
sc1,sc2,sc3,sc4 = list(),list(),list(),list()
ref = list()

for key in cap_dict:
  nest = cap_dict[key]
  can = nest[0]
  given = nest[1]
  for i in given:
    temp_str = i
    no_punct = ''
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    for char in temp_str:
      if char not in punctuations:
        no_punct = no_punct + char
    word_list = list(no_punct.split(" "))
    while('' in word_list):
      word_list.remove('') 

    ref.append(word_list)

  s1,s2,s3,s4 = bleu_score(can,ref)
  sc1.append(s1)
  sc2.append(s2)
  sc3.append(s3)
  sc4.append(s4)


avg_sc1 = sum(sc1)/len(sc1)
avg_sc2 = sum(sc2)/len(sc2)
avg_sc3 = sum(sc3)/len(sc3)
avg_sc4 = sum(sc4)/len(sc4)

print('BLEU-1: ', avg_sc4)
print('BLEU-2: ', avg_sc3)
print('BLEU-3: ', avg_sc2)
print('BLEU-4: ', avg_sc1)
```
